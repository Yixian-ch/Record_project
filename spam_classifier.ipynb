# import library

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

# load dataset

```python
spam_df = pd.read_csv("/content/drive/MyDrive/ml/spam.csv")
# inspect data
spam_df.groupby('Category').describe()
```

# Convert/Vectorize our Category data

```python
spam_df['spam'] = spam_df['Category'].apply(lambda x: 1 if x == 'spam' else 0)
spam_df
```

# Create trainning and testing data



```python
# from p(x|y) x are the count of our features, y is the label
x_train, x_test, y_train, y_test = train_test_split(spam_df.Message, spam_df.spam, train_size= 0.5) # half of Message will be splitted into test data
# def train_test_split(arrays, train_size) can split content from arrays, so a sample given, two results return
```

```python
x_train.describe() # count means total number of message
```

# Words tokenization

```python
cv = CountVectorizer()
# we need to count in our dataframe how many time each word occurs
x_train_count = cv.fit_transform(x_train.values) # create a bag-of-words BoW representation based on the vocabulary calculated before boW

##############################################################
# <2786x5983 sparse matrix of type '<class 'numpy.int64'>'####
# with 36823 stored elements in Compressed Sparse Row format>#
##############################################################
# 2786 rows,each row represent a message, there are 5983 columns, each represents a uniq word of vocabulary
# 36823 is the number of total words in trainning data
x_train_count.toarray()
```

## an example of the structure of BoW

```python
vocabulary = list(cv.vocabulary_) # no order
decode = []
encode_phrase = x_train_count.toarray()[0] # first row in our bag of words
for a in range(len(encode_phrase)):
    # then we use index a to get the word in vocabulary
    if encode_phrase[a] == 1:
        word = vocabulary[a]
        decode.append(word)
print(decode)
```

# Train model

```python
model = MultinomialNB()
model.fit(x_train_count, y_train)
```

# Testing model

```python
email_ham =  ["money money money"]
email_ham_count = cv.transform(email_ham) # no fit, we have already created our dictonary with trainning text
print(model.predict(email_ham_count))
```

```python
x_test_count = cv.transform(x_test)
model.score(x_test_count, y_test)
```
